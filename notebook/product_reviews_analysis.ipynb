```python
import pandas as pd

# load dataset from GitHub
url = "https://raw.githubusercontent.com/burloi-ship-it/ml-product-reviews-project/refs/heads/main/data/product_reviews_full.csv"

df = pd.read_csv(url)

# analysis
print(f"Number of rows: {df.shape[0]}")
print(f"\nFirst 5 rows: \n")
display(df.head(5))

# Show column data types and non-null counts
df.info()

# Count missing values per column
print(f"Missing values per column: \n")
print(df.isna().sum())
```

```python
# Visualize missing data with seaborn heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.heatmap(df.isna(), cbar=False, cmap="YlOrRd")
plt.title("Missing Values Heatmap")

plt.show()
```

```python
# count occurences of each sentiment label
sentiment_counts = df['sentiment'].value_counts()

# print counts
print(f"Sentiment distribution: \n")
print(sentiment_counts)
```

```python
# plot sentiment distribution
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.bar(sentiment_counts.index, sentiment_counts.values, color=['skyblue', 'salmon'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')

plt.show()
```

```python
# 1. Check the data type of the 'product_price' column
print(f"Data type of product_price column: ", df['product_price'].dtype)

# 2. Display the first 10 values from the column
print(f"First 10 values in product_price column: \n")
display(df['product_price'].head(10))

# 3. Show the 20 most frequent values in the column
print(f"Top 20 most frequent values in product_price column: \n")
print(df['product_price'].value_counts().head(20))

# 4. Check for known non-numeric text values
problematic_values = ['Free', 'Not Available', 'N/A', 'None', '-', 'free', 'unknown', 'Unavailable']
# Identify rows containing problematic values
mask_problematic = df['product_price'].astype(str).str.strip().isin(problematic_values)
df_problematic = df[mask_problematic]

print(f"\nFound {len(df_problematic)} rows with problematic textual values: ")
display(df_problematic[['product_price']].drop_duplicates())

# 5. Find and display some of the non-numeric values
price_clean = df['product_price'].astype(str).str.strip()
price_numeric = pd.to_numeric(price_clean, errors='coerce')

invalid_prices = df[price_numeric.isna()]
print(f"\nNumber of non-numeric prices: ", len(invalid_prices))
display(invalid_prices[['product_price']].drop_duplicates().head(20))
```

```python
# Drop all rows with missing values
df = df.dropna()

# Display the shape of dataset
print(f"New dataset shape: ", df.shape)

# Count missing values per column
print(f"Missing values per column: ")
print(df.isna().sum())
```

- Toate datele sunt convertite în string pentru a putea fi procesate textual.

- Ștergem simbolul valutar dacă există (de exemplu, $48.18 → 48.18).

- Eliminăm spațiile, simbolurile și tot ce nu este număr sau punct (de exemplu 1,199.99 → 1199.99).

- Eliminăm caracterele goale de la început și sfârșit.

- Convertim textul curățat în format numeric (float); valorile ce nu se pot converti devin NaN.

- Eliminăm rândurile cu valori NaN în coloana product_price, ștergând astfel prețurile invalide ce nu au putut fi parsate (dacă există).

- Afișăm statistici de bază (min, max, medie etc.) pentru a verifica succesul conversiei.

```python
# Step 1: Convert to string and remove the 'USD' prefix and any leading/trailing spaces
df['product_price_cleaned'] = (
    df['product_price']
    .astype(str)
    .str.replace(r'$', "", regex=True) # remove $
    .str.replace(r'[^\d.]', "", regex=True) # Remove all non-numeric characters except the dot
    .str.strip()
    )

# Step 2: Convert cleaned string to float
df['product_price'] = pd.to_numeric(df['product_price_cleaned'], errors='coerce')

# Step 3: Drop the temporary column
df = df.drop(columns=['product_price_cleaned'])

# Step 4: Drop any rows where conversion failed (still NaN)
df = df.dropna(subset=['product_price'])

# Step 5: Confirm result
print(f"Column type after parsing: ", df['product_price'].dtype)
print(f"\nPrice summary: ")
print(df['product_price'].describe())
```

```python
# Step 1: Convert all sentiment values to lowercase and strip extra spaces
df['sentiment'] = df['sentiment'].astype(str).str.lower().str.strip()

# Step 2: Show all unique values in sentiment column
print(f"Unique sentiments in sentiment column: \n")
print(df['sentiment'].value_counts())

# Step 3: Convert sentiment column in 'category'
df['sentiment'] = df['sentiment'].astype('category')
print(f"Sentiment column converted in: ", df['sentiment'].dtype)
```

```python
# Drop columns irrelevant for analysis
df = df.drop(columns=['review_uuid', 'product_name'])

# Preview remaining columns
print(f"Columns in dataframe: \n")
print(df.columns.tolist())
```

```python
# Price distribution related to sentiment
import seaborn as sns
import matplotlib.pyplot as plt

# Show summary stats grouped by sentiment
print(f"Price summary by sentiment: \n")
print(df.groupby('sentiment', observed=False)['product_price'].describe())

# Boxplot of prices by sentiment
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='sentiment', y='product_price')
plt.title("Product price by sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Product price")

plt.grid(True)
plt.show()
```
